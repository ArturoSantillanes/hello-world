Y la salida:

```python
w0 = 3
x0 = np.array([0.5, 0.1])
def dx(t, X) :
    x = X[0]
    dx = X[1]
    d2x = -w0**2*x
    return [dx, d2x]

t_s = [0, 20]
I = solve_ivp(dx, t_s, x0, max_step =0.1)
    
```
